```python
import numpy as np 
import pandas as pd 
```

```python
dfm = pd.read_csv('male_office_sitting4_scenario_newResults.csv')
dff = pd.read_csv('female_office_sitting4_scenario_newResults.csv')
```

```python
dff.head(10)
```

```python
dfm[dfm.columns[4]][:] = dfm[dfm.columns[4]][:]*1e-6
dff[dff.columns[4]][:] = dff[dff.columns[4]][:]*1e-6
dfm[dfm.columns[5]][:] = dfm[dfm.columns[5]][:]*2*1e-6*1.87/60/55.5
dff[dff.columns[5]][:] = dff[dff.columns[5]][:]*2*1e-6*1.87/60/55.5
dfm.rename(columns = {dfm.columns[4]:"MaleExpiratoryFlow(m3/s)"}, inplace = True)
dff.rename(columns = {dff.columns[4]:"FemaleExpiratoryFlow(m3/s)"}, inplace = True)
dfm.rename(columns = {dfm.columns[5]:"MaleCarbonDioxideProductionRate(kg/m3.s)"}, inplace = True)
dff.rename(columns = {dff.columns[5]:"FemaleCarbonDioxideProductionRate(kg/m3.s)"}, inplace = True)
#dfm.to_csv("inlet.csv", index=False, columns=[dfm.columns[0], dfm.columns[4], dff.columns[4], dfm.columns[5], dff.columns[5]])
data = {
    dfm.columns[0]: dfm[dfm.columns[0]],
    "MaleExpiratoryFlow(m3/s)": dfm[dfm.columns[4]],
    "FemaleExpiratoryFlow(m3/s)": dff[dff.columns[4]],
    "MaleCarbonDioxideProductionRate(kg/m3.s)": dfm[dfm.columns[5]],
    "FemaleCarbonDioxideProductionRate(kg/m3.s)": dff[dff.columns[5]]
}
df = pd.DataFrame(data)
df.to_csv("inlet.csv", index=False)
df2 = pd.read_csv('inlet.csv')
df2.head(10)
```
